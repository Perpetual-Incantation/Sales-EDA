```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel('Sales_data.xlsx')
df.head()
```

```python
df['Total price'] = df['Unit price'] * df['Quantity']
avg_prices = df.groupby('Branch')['Total price'].mean().reset_index()
avg_prices.columns = ['Branch', 'Average Price']
print(round(avg_prices,2))
```

```python
def barchart(x,hue,df,figsize=(5,4)):
    plt.figure(figsize=figsize)
    ax=sns.countplot(x=x, hue=hue, data=df)
    plt.title(f'Number of Customer by {x} and {hue}')
    plt.xlabel(x)
    plt.ylabel('Number of Customers')
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    plt.show()
    
def cir_chart(x):
    counts = df[x].value_counts()
    plt.pie(counts, autopct='%1.2f%%', startangle=80)
    plt.legend(counts.index)
    plt.title(f'Pie Chart of {x} Types')
    plt.show()
    
def costvs(vs, cost, hue=None, figsize=(4, 4)):
    costs = df.groupby([vs] + ([hue] if hue else []))[cost].sum().reset_index()
    plt.figure(figsize=figsize)
    if hue:
        sns.barplot(data=costs, x=vs, y=cost, hue=hue, errorbar=None)
    else:
        sns.barplot(data=costs, x=vs, y=cost, errorbar=None)
    plt.title(f'Total Cost Sum by {vs}')
    plt.xlabel(vs)
    plt.ylabel(f'Total Cost Sum by {vs}')
    ax = plt.gca()
    for container in ax.containers:
        ax.bar_label(container, label_type='edge', fmt='%.0f')
    if hue:
        plt.legend(title=hue)
    plt.show()
def RatingVs(Vs,Rating,df,figsize=(5,4)):
    plt.figure(figsize=figsize)
    sns.boxplot(x=Vs, y=Rating, data=df)
    plt.title(f'Rating by {Vs}')
    plt.xlabel(f'{Vs}')
    plt.ylabel(f'{Rating}')
    plt.show()
```
